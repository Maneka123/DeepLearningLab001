```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'new_df' is your dataframe containing the 4 features

g = sns.PairGrid(df)

# Upper triangle: Bivariate Histogram
g.map_upper(sns.histplot)

# Diagonal: Histogram with KDE
g.map_diag(sns.histplot, kde=True)

# Lower triangle: Bivariate KDE plot
g.map_lower(sns.kdeplot, cmap='Blues_d')

# Adjust layout
plt.tight_layout()
plt.show()
```

##  visualize the distribution histogram along with uniform and normal distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create 100000x1 array from exponential distribution
exponential_data = np.random.exponential(scale=1, size=100000)

# Create random data from uniform distribution for comparison
uniform_data = np.random.uniform(low=0, high=5, size=100000)

# Create random data from normal distribution for comparison
normal_data = np.random.normal(loc=2.5, scale=1, size=100000)

# Create a figure with subplots
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Plot histogram for exponential distribution
sns.histplot(exponential_data, kde=True, ax=axes[0])
axes[0].set_title('Exponential Distribution')

# Plot histogram for uniform distribution
sns.histplot(uniform_data, kde=True, ax=axes[1])
axes[1].set_title('Uniform Distribution')

# Plot histogram for normal distribution
sns.histplot(normal_data, kde=True, ax=axes[2])
axes[2].set_title('Normal Distribution')

# Adjust layout
plt.tight_layout()
plt.show()
```

## changing number of bins

```python
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

## plotting a 3D plot

```python
import matplotlib
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
#R = np.sqrt(X**2 + Y**2)
Z = X**2+Y**2

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```
